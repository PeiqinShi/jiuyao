```python
from cobra.io import read_sbml_model
from copy import deepcopy

model = read_sbml_model("./data/iYO844.xml")
```

```python
print("=== Model summary ===")
print("Reactions:", len(model.reactions))
print("Metabolites:", len(model.metabolites))
print("Genes:", len(model.genes))
print("Objective:", model.objective.expression)
print()

# =========================
# 2. 查看默认培养基
# =========================
print("=== Default medium (allowed uptake) ===")
for rxn in model.exchanges:
    if rxn.lower_bound < 0:
        print(f"{rxn.id:20s} {rxn.lower_bound:12.4f}   {rxn.name}")
print()

print("=== model.medium ===")
print(model.medium)
print()

# =========================
# 3. 列出所有 exchange reactions
# =========================
print("=== All exchange reactions ===")
for rxn in model.exchanges:
    print(f"{rxn.id:20s} {rxn.name}")
print()

# =========================
# 4. 搜索关键 exchange IDs
# =========================
print("=== Search possible maltose exchange IDs ===")
for rxn in model.exchanges:
    text = (rxn.id + " " + rxn.name).lower()
    if "malt" in text:
        print(f"{rxn.id:20s} {rxn.name}")
print()

print("=== Search possible starch/dextrin exchange IDs ===")
for rxn in model.exchanges:
    text = (rxn.id + " " + rxn.name).lower()
    if "starch" in text or "dextrin" in text or "amylase" in text:
        print(f"{rxn.id:20s} {rxn.name}")
print()

print("=== Search key exchange IDs ===")
keywords = ["glc", "glucose", "o2", "oxygen", "nh4", "ammon", "pi", "phosphate", "so4", "sulfate", "fe3", "iron"]
for rxn in model.exchanges:
    text = (rxn.id + " " + rxn.name).lower()
    if any(k in text for k in keywords):
        print(f"{rxn.id:20s} {rxn.name}")
print()

# =========================
# 5. 工具函数：设置最小培养基
# =========================
def set_minimal_medium(model, carbon_source_id, carbon_uptake=10.0, oxygen_uptake=20.0):
    """
    构建标准最小培养基：
    - 关闭所有摄取
    - 只打开一个碳源 + 必要无机盐/离子/铁
    - model.medium 里用正数表示允许摄取上限
    """
    medium = {rxn.id: 0.0 for rxn in model.exchanges}

    candidate_medium = {
        carbon_source_id: carbon_uptake,
        "EX_o2_e": oxygen_uptake,
        "EX_nh4_e": 10.0,
        "EX_pi_e": 10.0,
        "EX_so4_e": 10.0,
        "EX_fe3_e": 1000.0,
        "EX_h2o_e": 1000.0,
        "EX_h_e": 1000.0,
        "EX_k_e": 1000.0,
        "EX_na1_e": 1000.0,
        "EX_mg2_e": 1000.0,
        "EX_ca2_e": 1000.0,
    }

    exchange_ids = {rxn.id for rxn in model.exchanges}

    for rid, val in candidate_medium.items():
        if rid in exchange_ids:
            medium[rid] = val
        else:
            print(f"[Warning] {rid} not found in model exchanges")

    model.medium = medium


# =========================
# 6. 工具函数：打印培养基
# =========================
def print_medium(model):
    print("Medium:")
    for k, v in model.medium.items():
        if v > 0:
            print(f"{k:20s} {v:12.4f}")
    print()


# =========================
# 7. 工具函数：打印实际 uptake / secretion
# =========================
def print_fluxes(solution, model, threshold=1e-6):
    print("=== Actual uptake fluxes ===")
    for rxn in model.exchanges:
        flux = solution.fluxes[rxn.id]
        if flux < -threshold:
            print(f"{rxn.id:20s} {flux:12.6f}   {rxn.name}")
    print()

    print("=== Actual secretion fluxes ===")
    for rxn in model.exchanges:
        flux = solution.fluxes[rxn.id]
        if flux > threshold:
            print(f"{rxn.id:20s} {flux:12.6f}   {rxn.name}")
    print()


# =========================
# 8. 工具函数：运行一个条件
# =========================
def run_condition(base_model, condition_name, carbon_source_id, carbon_uptake, oxygen_uptake):
    """
    复制模型 -> 设置培养基 -> 求解 -> 输出结果
    """
    test_model = base_model.copy()

    print("=" * 70)
    print(f"Condition: {condition_name}")
    print("=" * 70)

    set_minimal_medium(
        test_model,
        carbon_source_id=carbon_source_id,
        carbon_uptake=carbon_uptake,
        oxygen_uptake=oxygen_uptake
    )

    print_medium(test_model)

    solution = test_model.optimize()

    print("Status:", solution.status)
    print("Growth rate:", solution.objective_value)
    print()

    print_fluxes(solution, test_model)

    return solution


# =========================
# 9. 先确认关键碳源 reaction 是否存在
# =========================
exchange_ids = {rxn.id for rxn in model.exchanges}

glucose_id = "EX_glc__D_e"
maltose_id = "EX_malt_e"

print("=== Carbon source ID check ===")
print("Glucose exists:", glucose_id in exchange_ids)
print("Maltose exists:", maltose_id in exchange_ids)
print()

# =========================
# 10. 跑三个条件
# =========================

# 条件1：葡萄糖 + 有氧
sol_glc_aerobic = run_condition(
    base_model=model,
    condition_name="glucose + aerobic",
    carbon_source_id=glucose_id,
    carbon_uptake=10.0,
    oxygen_uptake=20.0
)

# 条件2：葡萄糖 + 限氧
sol_glc_low_o2 = run_condition(
    base_model=model,
    condition_name="glucose + oxygen-limited",
    carbon_source_id=glucose_id,
    carbon_uptake=10.0,
    oxygen_uptake=5.0
)

# 条件3：麦芽糖 + 有氧
if maltose_id in exchange_ids:
    sol_malt_aerobic = run_condition(
        base_model=model,
        condition_name="maltose + aerobic",
        carbon_source_id=maltose_id,
        carbon_uptake=10.0,
        oxygen_uptake=20.0
    )
else:
    sol_malt_aerobic = None
    print("[Warning] Maltose exchange ID not found, skipped maltose condition.")
    print()

# =========================
# 11. 汇总结果
# =========================
print("=" * 70)
print("Summary")
print("=" * 70)
print(f"{'Condition':30s} {'Status':>12s} {'Growth rate':>15s}")

print(f"{'glucose + aerobic':30s} {sol_glc_aerobic.status:>12s} {sol_glc_aerobic.objective_value:15.6f}")
print(f"{'glucose + oxygen-limited':30s} {sol_glc_low_o2.status:>12s} {sol_glc_low_o2.objective_value:15.6f}")

if sol_malt_aerobic is not None:
    print(f"{'maltose + aerobic':30s} {sol_malt_aerobic.status:>12s} {sol_malt_aerobic.objective_value:15.6f}")
else:
    print(f"{'maltose + aerobic':30s} {'skipped':>12s} {'NA':>15s}")

print()

# =========================
# 12. 可选：把结果整理成字典，后面方便画表
# =========================
results = {
    "glucose_aerobic": {
        "status": sol_glc_aerobic.status,
        "growth_rate": sol_glc_aerobic.objective_value
    },
    "glucose_low_o2": {
        "status": sol_glc_low_o2.status,
        "growth_rate": sol_glc_low_o2.objective_value
    }
}

if sol_malt_aerobic is not None:
    results["maltose_aerobic"] = {
        "status": sol_malt_aerobic.status,
        "growth_rate": sol_malt_aerobic.objective_value
    }

print("=== Results dict ===")
print(results)
```

```python
import pandas as pd
from cobra.flux_analysis import pfba

# =========================================================
# 1. 条件运行函数：返回 model + optimize解 + pfba解
# =========================================================
def run_condition_with_pfba(base_model, condition_name, carbon_source_id, carbon_uptake, oxygen_uptake):
    test_model = base_model.copy()

    print("=" * 70)
    print(f"Condition: {condition_name}")
    print("=" * 70)

    set_minimal_medium(
        test_model,
        carbon_source_id=carbon_source_id,
        carbon_uptake=carbon_uptake,
        oxygen_uptake=oxygen_uptake
    )

    print_medium(test_model)

    sol_opt = test_model.optimize()
    print("Status:", sol_opt.status)
    print("Growth rate:", sol_opt.objective_value)
    print()

    print_fluxes(sol_opt, test_model)

    # 再跑 pFBA，后面提关键通路更稳
    sol_pfba = pfba(test_model)

    return {
        "condition": condition_name,
        "model": test_model,
        "opt_solution": sol_opt,
        "pfba_solution": sol_pfba
    }


# =========================================================
# 2. 定义要跑的条件
# =========================================================
conditions = [
    {
        "condition_name": "glucose_aerobic",
        "carbon_source_id": "EX_glc__D_e",
        "carbon_uptake": 10.0,
        "oxygen_uptake": 20.0
    },
    {
        "condition_name": "glucose_low_o2",
        "carbon_source_id": "EX_glc__D_e",
        "carbon_uptake": 10.0,
        "oxygen_uptake": 5.0
    },
    {
        "condition_name": "maltose_aerobic",
        "carbon_source_id": "EX_malt_e",
        "carbon_uptake": 10.0,
        "oxygen_uptake": 20.0
    },
    {
        "condition_name": "dextrin_aerobic",
        "carbon_source_id": "EX_dextrin_e",
        "carbon_uptake": 10.0,
        "oxygen_uptake": 20.0
    },
    {
        "condition_name": "starch_aerobic",
        "carbon_source_id": "EX_starch_e",
        "carbon_uptake": 10.0,
        "oxygen_uptake": 20.0
    }
]


# =========================================================
# 3. 批量运行
# =========================================================
bs_results = []

exchange_ids = {rxn.id for rxn in model.exchanges}

for c in conditions:
    if c["carbon_source_id"] not in exchange_ids:
        print(f"[Skip] {c['condition_name']} because {c['carbon_source_id']} not found.")
        continue
    
    res = run_condition_with_pfba(
        base_model=model,
        condition_name=c["condition_name"],
        carbon_source_id=c["carbon_source_id"],
        carbon_uptake=c["carbon_uptake"],
        oxygen_uptake=c["oxygen_uptake"]
    )
    bs_results.append(res)


# =========================================================
# 4. 导出 exchange flux summary
# =========================================================
def classify_metabolite(rxn_id, rxn_name):
    text = f"{rxn_id} {rxn_name}".lower()
    
    if any(k in text for k in ["glc", "glucose", "starch", "dextrin", "malt", "glycogen", "sucrose", "fru", "carbon"]):
        return "carbon"
    elif any(k in text for k in ["nh4", "ammon", "amino", "arg", "asp", "glu", "lys", "his", "ser", "gly", "ala", "val", "leu", "ile", "pro", "met", "asn", "gln"]):
        return "nitrogen"
    elif any(k in text for k in ["o2", "oxygen", "co2"]):
        return "gas"
    elif any(k in text for k in ["ac", "acetate", "lact", "form", "etoh", "succ", "pyr", "glyclt"]):
        return "fermentation_product"
    elif any(k in text for k in ["pi", "so4", "fe", "mg", "ca", "na", "k", "h2o", "h+"]):
        return "ion_or_inorganic"
    else:
        return "other"


exchange_rows = []

for res in bs_results:
    cond = res["condition"]
    m = res["model"]
    sol = res["opt_solution"]
    
    for rxn in m.exchanges:
        flux = sol.fluxes[rxn.id]
        if abs(flux) > 1e-6:
            direction = "uptake" if flux < 0 else "secretion"
            exchange_rows.append({
                "species": "B_subtilis",
                "condition": cond,
                "rxn_id": rxn.id,
                "rxn_name": rxn.name,
                "flux": float(flux),
                "direction": direction,
                "metabolite_class": classify_metabolite(rxn.id, rxn.name)
            })

exchange_df = pd.DataFrame(exchange_rows)
exchange_df.to_csv("B_subtilis_exchange_flux_summary.csv", index=False)

print("Saved: B_subtilis_exchange_flux_summary.csv")
print(exchange_df.head())


# =========================================================
# 5. 导出关键通路 flux
#    这里优先抓“你论文会用到”的关键反应
# =========================================================
keywords = [
    "amyl", "dextrin", "malt", "glc", "glycol", "pyr", "lact", "acet", "form",
    "tca", "cit", "accoa", "biomass", "starch", "suc", "g6p", "pep", "pk"
]

def is_key_reaction(rxn):
    text = f"{rxn.id} {rxn.name} {rxn.reaction}".lower()
    return any(k in text for k in keywords)

pathway_rows = []

for res in bs_results:
    cond = res["condition"]
    m = res["model"]
    sol = res["pfba_solution"]   # 关键通路用 pFBA
    
    for rxn in m.reactions:
        if is_key_reaction(rxn):
            flux = sol.fluxes[rxn.id]
            if abs(flux) > 1e-6:
                pathway_rows.append({
                    "species": "B_subtilis",
                    "condition": cond,
                    "reaction_id": rxn.id,
                    "reaction_name": rxn.name,
                    "reaction_equation": rxn.reaction,
                    "flux": float(flux)
                })

pathway_df = pd.DataFrame(pathway_rows)
pathway_df = pathway_df.sort_values(["condition", "flux"], ascending=[True, False])
pathway_df.to_csv("B_subtilis_key_pathway_flux.csv", index=False)

print("Saved: B_subtilis_key_pathway_flux.csv")
print(pathway_df.head())


# =========================================================
# 6. 额外导出生长汇总表（建议一起留着）
# =========================================================
growth_rows = []

for res in bs_results:
    growth_rows.append({
        "species": "B_subtilis",
        "condition": res["condition"],
        "status": res["opt_solution"].status,
        "growth_rate": float(res["opt_solution"].objective_value)
    })

growth_df = pd.DataFrame(growth_rows)
growth_df.to_csv("B_subtilis_growth_summary.csv", index=False)

print("Saved: B_subtilis_growth_summary.csv")
print(growth_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib as mpl

mpl.rcParams['pdf.fonttype'] = 42   # ⭐关键
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['font.family'] = 'Arial'
# ==============================
# 1. 读取数据
# ==============================
df = pd.read_csv("./data/B_subtilis_exchange_flux_summary.csv")

# ==============================
# 2. 过滤：保留所有非0 exchange，但去掉不想展示的无机离子/水
# ==============================
exclude_patterns = [
    "EX_fe3_e",
    "EX_ca2_e",
    "EX_so4_e",
    "EX_k_e",
    "EX_mg2_e",
    "EX_pi_e",
    "EX_h2o_e"
]

df_plot = df[~df["rxn_id"].isin(exclude_patterns)].copy()

# ==============================
# 3. 美化名称
#    没写到映射表里的，会自动用 rxn_id
# ==============================
rxn_name_map = {
    "EX_glc__D_e": "Glucose",
    "EX_malt_e": "Maltose",
    "EX_dextrin_e": "Dextrin",
    "EX_starch_e": "Starch",
    "EX_ac_e": "Acetate",
    "EX_lac__L_e": "L-Lactate",
    "EX_for_e": "Formate",
    "EX_co2_e": "CO2",
    "EX_h_e": "H+",
    "EX_o2_e": "O2",
    "EX_nh4_e": "NH4+",
    "EX_glyclt_e": "Glycolate"
}

df_plot["Metabolite"] = df_plot["rxn_id"].map(rxn_name_map).fillna(df_plot["rxn_id"])

# ==============================
# 4. 条件顺序
# ==============================
condition_order = [
    "glucose_aerobic",
    "glucose_low_o2",
    "maltose_aerobic",
    "dextrin_aerobic",
    "starch_aerobic"
]

condition_name_map = {
    "glucose_aerobic": "Glucose\nAerobic",
    "glucose_low_o2": "Glucose\nLow O2",
    "maltose_aerobic": "Maltose\nAerobic",
    "dextrin_aerobic": "Dextrin\nAerobic",
    "starch_aerobic": "Starch\nAerobic"
}

df_plot["Condition"] = df_plot["condition"].map(condition_name_map)

# ==============================
# 5. 自动生成代谢物顺序
#    这里按第一次出现顺序；你也可以手动指定
# ==============================
metabolite_order = list(dict.fromkeys(df_plot["Metabolite"].tolist()))

# ==============================
# 6. 方向和气泡大小
# ==============================
df_plot["Direction"] = df_plot["flux"].apply(lambda x: "Import" if x < 0 else "Export")

# 限制 flux 到 0-50，并严格用它控制气泡大小
df_plot["AbsFlux"] = df_plot["flux"].abs().clip(upper=50)

min_size = 20   # ⭐最小气泡大小（关键参数）
df_plot["s"] = df_plot["AbsFlux"] * size_scale + min_size

# ==============================
# 7. 坐标映射
# ==============================
x_labels = [condition_name_map[c] for c in condition_order if c in df_plot["condition"].unique()]
x_map = {name: i for i, name in enumerate(x_labels)}
y_map = {name: i for i, name in enumerate(metabolite_order)}

df_plot["x"] = df_plot["Condition"].map(x_map)
df_plot["y"] = df_plot["Metabolite"].map(y_map)

# 去掉映射失败的行（保险）
df_plot = df_plot.dropna(subset=["x", "y"]).copy()

# ==============================
# 8. 配色
# ==============================
palette = {
    "Export": "#52C7C7",
    "Import": "#9ACD66"
}

# ==============================
# 9. 作图
# ==============================
fig, ax = plt.subplots(figsize=(5, 6.2))

for direction in ["Export", "Import"]:
    sub = df_plot[df_plot["Direction"] == direction]
    ax.scatter(
        sub["x"],
        sub["y"],
        s=sub["s"],
        c=palette[direction],
        alpha=0.95,
        edgecolors="none",
        linewidths=0,
        label=direction
    )

# ==============================
# 10. 坐标轴设置
# ==============================
ax.set_xticks(range(len(x_labels)))
ax.set_xticklabels(x_labels, fontsize=10)

ax.set_yticks(range(len(metabolite_order)))
ax.set_yticklabels(metabolite_order, fontsize=10)

ax.invert_yaxis()
ax.set_xlabel("")
ax.set_ylabel("Exchanged metabolite", fontsize=12)
ax.set_title("Exchange flux profiles of Bacillus subtilis", fontsize=13, pad=12)

# 去边框、去网格、透明背景
for spine in ax.spines.values():
    spine.set_visible(False)

ax.grid(False)
ax.set_facecolor("none")
fig.patch.set_alpha(0)

# ==============================
# 11. 图例
# ==============================
# Direction legend
export_patch = mlines.Line2D([], [], color=palette["Export"], marker="o", linestyle="None",
                             markersize=8, label="Export")
import_patch = mlines.Line2D([], [], color=palette["Import"], marker="o", linestyle="None",
                             markersize=8, label="Import")

legend1 = ax.legend(
    handles=[export_patch, import_patch],
    title="Direction",
    loc="center left",
    bbox_to_anchor=(1.02, 0.70),
    frameon=False
)
ax.add_artist(legend1)

# Flux legend（与主图严格对应）
legend_flux = [10, 20, 30, 40]
size_handles = [
    ax.scatter([], [], s=flux * size_scale, color="gray", alpha=0.8,
               edgecolors="none", linewidths=0)
    for flux in legend_flux
]

legend2 = ax.legend(
    handles=size_handles,
    labels=[str(x) for x in legend_flux],
    title="Flux",
    loc="center left",
    bbox_to_anchor=(1.02, 0.36),
    frameon=False,
    labelspacing=1.8
)

# ==============================
# 12. 输出
# ==============================
plt.tight_layout()
plt.savefig(
    "B_subtilis_exchange_bubbleplot.pdf",
    format="pdf",
    bbox_inches="tight"
)
plt.show()
```

```python
# ==============================
# 2. 导出代谢物信息
# ==============================
metabolite_rows = []

for met in model.metabolites:
    metabolite_rows.append({
        "metabolite_id": met.id,
        "metabolite_name": met.name,
        "compartment": met.compartment,
        "formula": getattr(met, "formula", None),
        "charge": getattr(met, "charge", None)
    })

metabolites_df = pd.DataFrame(metabolite_rows)
metabolites_df.to_csv("./data/B_subtilis_metabolites.csv", index=False)

print("Saved: ./data/B_subtilis_metabolites.csv")
print(metabolites_df.head())


# ==============================
# 3. 导出反应信息
# ==============================
reaction_rows = []

for rxn in model.reactions:
    # subsystem 有些模型没有，先安全获取
    subsystem = None
    if hasattr(rxn, "subsystem"):
        subsystem = rxn.subsystem

    # 有些模型把通路信息放在 annotation 里
    if not subsystem and hasattr(rxn, "annotation"):
        subsystem = rxn.annotation.get("subsystem", None)

    reaction_rows.append({
        "reaction_id": rxn.id,
        "reaction_name": rxn.name,
        "reaction_equation": rxn.reaction,
        "lower_bound": rxn.lower_bound,
        "upper_bound": rxn.upper_bound,
        "objective_coefficient": rxn.objective_coefficient,
        "gene_reaction_rule": rxn.gene_reaction_rule,
        "subsystem": subsystem
    })

reactions_df = pd.DataFrame(reaction_rows)
reactions_df.to_csv("./data/B_subtilis_reactions.csv", index=False)

print("Saved: ./data/B_subtilis_reactions.csv")
print(reactions_df.head())


# ==============================
# 4. 导出“通路/子系统-反应”对应表
# ==============================
pathway_df = reactions_df[["reaction_id", "reaction_name", "subsystem"]].copy()
pathway_df.to_csv("./data/B_subtilis_reaction_pathways.csv", index=False)

print("Saved: ./data/B_subtilis_reaction_pathways.csv")
print(pathway_df.head())


# ==============================
# 5. 导出每个通路包含多少反应（汇总表）
# ==============================
pathway_summary_df = (
    reactions_df
    .assign(subsystem=reactions_df["subsystem"].fillna("Unknown"))
    .groupby("subsystem", as_index=False)
    .agg(
        reaction_count=("reaction_id", "count")
    )
    .sort_values("reaction_count", ascending=False)
)

pathway_summary_df.to_csv("./data/B_subtilis_pathway_summary.csv", index=False)

print("Saved: ./data/B_subtilis_pathway_summary.csv")
print(pathway_summary_df.head(20))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# 1. 数据
# ==============================
data = {
    "species": ["B. subtilis", "S. cerevisiae", "R. microsporus"],
    "Sugars": [76, 104, 85],
    "Amino acids": [59, 65, 62],
    "Vitamins": [29, 37, 31],
    "Salts & minerals": [36, 41, 39],
    "Organic acids": [79, 88, 83],
    "Alcohols": [38, 53, 45],
    "Nucleotides": [71, 75, 73]
}

df = pd.DataFrame(data)

# 转长格式
df_long = df.melt(id_vars="species", var_name="Category", value_name="Count")

# ==============================
# 2. 顺序
# ==============================
species_order = ["B. subtilis", "S. cerevisiae", "R. microsporus"]
category_order = [
    "Sugars",
    "Amino acids",
    "Vitamins",
    "Salts & minerals",
    "Organic acids",
    "Alcohols",
    "Nucleotides"
]

df_long["species"] = pd.Categorical(df_long["species"], categories=species_order, ordered=True)
df_long["Category"] = pd.Categorical(df_long["Category"], categories=category_order, ordered=True)

# ==============================
# 3. 颜色（按类别固定）
# ==============================
palette = {
    "Sugars": "#F4A261",            # 橙
    "Amino acids": "#8BC34A",       # 绿
    "Vitamins": "#2EC4B6",          # 青绿
    "Salts & minerals": "#E76F51",  # 红橙
    "Organic acids": "#5DADE2",     # 蓝
    "Alcohols": "#D3D3D3",          # 灰
    "Nucleotides": "#B39DDB"        # 淡紫
}

# ==============================
# 4. 气泡大小
# ==============================
sns.set_style("white")

# 统一缩小后的气泡范围
min_size, max_size = 30, 500

# 使用 Count 做大小映射
df_long["BubbleSize"] = df_long["Count"]

# 为了让右侧 Count 图例和左侧气泡完全一致，定义同一套映射函数
min_val = df_long["Count"].min()
max_val = df_long["Count"].max()

def size_map(x):
    return min_size + (x - min_val) / (max_val - min_val) * (max_size - min_size)

# ==============================
# 5. 作图
# ==============================
plt.figure(figsize=(7, 3.2))

ax = sns.scatterplot(
    data=df_long,
    x="Category",
    y="species",
    size="BubbleSize",
    hue="Category",
    sizes=(min_size, max_size),
    palette=palette,
    edgecolor="none",
    alpha=0.95,
    legend=False,
    zorder=3
)

# 防止气泡被边界裁剪
for coll in ax.collections:
    coll.set_clip_on(False)

# ==============================
# 6. 美化
# ==============================
ax.set_xlabel("")
ax.set_ylabel("Species", fontsize=12)
ax.set_title("Metabolite class distribution", fontsize=13, pad=14)

# x轴放到上方，标签45度
ax.xaxis.tick_top()
ax.xaxis.set_label_position("top")
ax.tick_params(axis="x", bottom=False, top=True, length=0)
ax.tick_params(axis="y", length=0)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="left", fontsize=10)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=10)

# 边界留白，避免第一行圆圈显示不全
ax.set_xlim(-0.4, len(category_order) - 0.6)
ax.set_ylim(len(species_order) - 0.4, -0.6)

# 去掉边框
for spine in ax.spines.values():
    spine.set_visible(False)

# ==============================
# 7. 气泡大小图例（与左图大小严格对应）
# ==============================
size_levels = [30, 50, 75, 100]

size_handles = [
    plt.scatter([], [], s=size_map(s), color="gray", alpha=0.7, edgecolors="none")
    for s in size_levels
]

ax.legend(
    handles=size_handles,
    labels=[str(s) for s in size_levels],
    title="Count",
    loc="center left",
    bbox_to_anchor=(1.02, 0.5),
    frameon=False,
    labelspacing=1.5,
    borderaxespad=0.0
)

# ==============================
# 8. 保存 / 显示
# ==============================
plt.tight_layout()

# 保存为高分辨率图片
plt.savefig("metabolite_class_bubbleplot_updated.pdf", dpi=300, bbox_inches="tight", transparent=True)

plt.show()
```
